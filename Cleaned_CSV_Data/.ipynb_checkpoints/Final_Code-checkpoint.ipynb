```python
# Add pandas as dependency

import pandas as pd

import os

import matplotlib.pyplot as plt
```

```python
# Files to load

netflix_data_to_load = "Resources/n_movies.csv"
```

```python
# Read the data and store it in Pandas DataFrame

netflix_data_df = pd.read_csv(netflix_data_to_load)

netflix_data_df.head()
```

```python
netflix_data_df.columns
```

```python
netflix_data_df.reset_index()
```

```python
print(netflix_data_df['genre'])
```

```python
netflix_data_df = netflix_data_df[['title', 'year', 'certificate', 'duration', 'genre', 'rating']]

netflix_data_df
```

```python
netflix_data_df['year'] = netflix_data_df['year'].str.replace('(', '')

netflix_data_df
```

```python
netflix_data_df['year'] = netflix_data_df['year'].str.replace(')', '')

netflix_data_df.head()
```

```python
netflix_data_df['year released'] = netflix_data_df['year'].str[0:4]

netflix_data_df['year released']
```

```python
netflix_data_df['duration_min'] = netflix_data_df['duration'].str[0:2]

netflix_data_df['duration_min']
```

```python
new = netflix_data_df['genre'].str.split(",", n = 3, expand = True)

netflix_data_df['genre_1'] = new[0]
netflix_data_df['genre_2'] = new[1]
netflix_data_df['genre_3'] = new[2]

netflix_data_df
```

```python
netflix_data_df = netflix_data_df[['title', 'year released', 'certificate', 'duration_min', 'genre_1', 'genre_2', 'genre_3', 'rating']]

netflix_data_df.head()
```

```python
netflix_data_df_drop = netflix_data_df.dropna()

netflix_data_df_drop
```

```python
new_movie_df = netflix_data_df_drop.sort_values(by = ['year released'])

new_movie_df
```

```python
new_movie_df.reset_index()
```

```python
new_movie_df.drop(new_movie_df.tail(243).index, 
                     inplace = True)

new_movie_df
```

```python
new_movie_df.columns
```

```python
new_movie_df['year released'] = new_movie_df['year released'].astype('int')
```

```python
x = new_movie_df.loc[new_movie_df['year released'] > 2011]
y = new_movie_df.groupby(['year released'])['rating'].mean()

plt.bar(x, y)
plt.xlabel('year released')
plt.ylabel('rating')
plt.show
```

```python
new_movie_df['rating category'] = pd.cut(x = new_movie_df['rating'],
                                       bins = [1, 4, 7, 10],
                                       labels = ['Bad', 'Good', 'Excellent'])

new_movie_df
```

```python
df_copy = new_movie_df.copy()
```

```python
df_copy.drop('title', axis = 1, inplace = True)
```

```python
y = df_copy["rating category"]
#X = df_copy.drop(['rating', 'rating category'], axis = 1).values
X = df_copy.drop(columns = "rating category")
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(X["certificate"])
```

```python
X["certificate"] = le.transform(X["certificate"])
```

```python
X["certificate"]
```

```python
le.inverse_transform(X["certificate"])
```

```python
le.fit(X["genre_1"])
X["genre_1"] = le.transform(X["genre_1"])
```

```python
X["genre_1"]
```

```python
le.inverse_transform(X["genre_1"])
```

```python
le.fit(X["genre_2"])
X["genre_2"] = le.transform(X["genre_2"])
```

```python
X["genre_2"]
```

```python
le.fit(X["genre_3"])
X["genre_3"] = le.transform(X["genre_3"])
```

```python
X["genre_3"]
```

```python
le.inverse_transform(X["genre_2"])
```

```python
le.inverse_transform(X["genre_3"])
```

```python
#from sklearn.model_selection import train_test_split

#X_train, X_test, y_train, y_test = train_test_split(X,
                                                   #y,
                                                   #random_state = 1,
                                                   #stratify = y)
            
            
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)

X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver = 'lbfgs',
                               max_iter = 200,
                               random_state = 1)
        
#from sklearn.linear_model import LogisticRegression
#lr = LogisticRegression()
#lr.fit(X_train, y_train)
```

```python
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```
